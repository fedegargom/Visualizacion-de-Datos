```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/weight-height.csv')
df.head()
```

```python
df.corr()
```

```python
x = df['Height']
y = df['Weight']
x = x.to_frame()
y = y.to_frame()
```

```python
pip install statsmodels
```

```python
pip install sklearn
```

```python
import statsmodels.api as sm
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
lm
```

```python
model_lm = lm.fit(y,x)
```

```python
print('El intercepto es \t:', model_lm.intercept_)
print('El coefficiente es \t:', model_lm.coef_)
print('El valor de R^2 es \t:', model_lm.score(y,x))
```

```python
model_ols = sm.OLS.from_formula('x~y', data = df)
result = model_ols.fit()
result.summary()
```

```python
numeros = ([200, 201, 94])
numeros = pd.DataFrame(numeros)
numeros
```

```python
y = model_lm.predict(numeros)
y = pd.DataFrame(y)
y
```
