```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
df.columns
```

```python
vars = ['SMQ020', 'RIAGENDR', 'RIDAGEYR','BMXWT', 'BMXHT', 'BMXBMI', 'BMXWAIST', 'HIQ210']
```

```python
df = df[vars].dropna()
df
```

```python
df.corr()
```

## BMXWT = b0 + b1*RIDAGEYR

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR', data = df)
result = model.fit()
result.summary()
```

```python
df['RiaGENDR'] = df.RIAGENDR.replace({1:'Male', 2:'Female'})
df
```

### BMXWT = b0 + b1*RIDAGEYR + b2*RiaGENDR
### BMXWT = b0 + b1*RIDAGEYR + b2*(Male+Female)
### BMXWT = 77.5710 -0.0228b1*RIDAGEYR + 10.5371*Male
### BMXWT = 77.5710 -0.0228b1*RIDAGEYR + *Female

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR + RiaGENDR', data = df)
result = model.fit()
result.summary()
```
