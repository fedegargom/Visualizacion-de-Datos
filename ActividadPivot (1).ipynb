## A3 PIVOT TABLES PROBLEMA 1

```python
import pandas as pd
import numpy as np
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['desc_ent','entidad'],
                          values = ['peso'],
                          aggfunc = ['mean','median'],
                          columns = 'sexo',
                          margins = True
                         )
my_pivot
```
