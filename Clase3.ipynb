## Clase3: Tablas Pivote

```python
import pandas as pd
```

```python
import numpy as np
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['STNAME','CTYNAME'],
                          values = ['BIRTHS2010','BIRTHS2011'],
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'median',
                          columns = 'SUMLEV',
                          margins = True
                         )
my_pivot
```

## Ejemplo con otra Base de Datos

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/airline_data.csv',low_memory=False)
df.head()
```

```python
df.shape
```

```python
columnz = df.columns.tolist()
columnz.sort()
columnz
```

```python
delay_pivot = pd.pivot_table(df,
                             index = ['Year','Quarter'],
                             values = 'DepDelay',
                             aggfunc = 'sum',
                          #aggfunc = 'median',
                             columns = 'DepartureDelayGroups',
                             margins = True
                         )
delay_pivot
```

```python
df[['DepDelay','DepartureDelayGroups']]
```
