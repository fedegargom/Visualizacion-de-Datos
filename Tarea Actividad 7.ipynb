```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
cintura = df['cintura'].plot(kind = 'hist', bins = 15)
```

```python
sns.lmplot(x = 'peso',
           y= 'cintura',
           hue = 'sexo',
           data = df,
           fit_reg = False, 
          )
```

```python
hombre = df.loc[df['sexo'] == 1]
hombre.head()
```

```python
sns.lmplot(x = 'peso', 
           y = 'cintura', 
           data = hombre,
           fit_reg = False, 
           height=5, aspect=2
          )
```
